# **NUMPY 함수 작성**


*   기본적인 numpy array 작성
*   array 연산 진행



```python
#리스트 생성
import numpy as np
ex = [1,9,-50,100]

#numpy array 변환 및 연산
ex_1 = np.array(ex)
ex_2 = ex_1 + 70
print(ex_1)
print(ex_2)
type(ex_1)
```

array간 연산

```python
diff = ex_2-ex_1
sum = ex_1+ex_2
multi = ex_1*ex_2
div = ex_2/ex_1
print(diff)
print(sum)
print(multi)
print(div)
```

2차 배열 만들기

```python
#2차 배열 생성
ex_2d = [diff, sum, multi, div]
print(ex_2d)
#2차 numpy array 생성 및 data type 변경
ex_2d_arr = np.array((ex_2d), dtype = np.int64)
print(ex_2d_arr)
print(ex_2d_arr[1])
print(ex_2d_arr[0][2])
```

```python
#4*4 무작위 행렬 생성
arr1 = np.random.randint(0,20, size=(4,4))
print(arr1)
```

```python
#arr1 전치행렬 생성
arr2 = arr1.transpose()
print(arr2)
```

행렬곱 계산

```python
arr1@arr2
```

주요 통계값 확인

```python
print(arr2.max())
print(arr2.min())
print(arr2.std())
print(arr2.var())
print(arr2.mean())
print(arr2.sum())
print(arr2.cumsum())
```

특정 위치 출력

```python
print(arr2[:3])
print(arr2[:1])
print(arr2[1:3])
print(arr2[:])
```

```python
arr2.shape
```

```python
arr3=np.linspace(1,10, num=19)
print(arr3)
```

차원 변경하기

```python
#1차원 추가하기
arr4 = np.expand_dims(arr2,1)
print(arr4)
print(arr4.shape)
#차원 삭제
arr4=arr4.squeeze(1)
print(arr4)
print(arr4.shape)
#형태 변경
arr4 = np.arange(16).reshape(2,8)
print(arr4)
print(arr4.shape)
#순서 변경
arr5 = arr4.transpose(1,0)
print(arr5)
print(arr5.shape)
```

# Pandas 함수 작성


*   dictionary 및 dataframe 활용
*   colab 연동
*   외부파일 읽기 및 쓰기



```python
#pandas 선언
import pandas as pd
#dictionary 생성
co = pd.Series({'이름':'홍길동','키':174,'몸무게':70})
print(co,type(co))
```

```python
#dataframe 생성
cop=[
     {'이름':'홍길동','키':174,'몸무게':70},
     {'이름':'홍일점','키':164,'몸무게':60},
     {'이름':'홍진규','키':184,'몸무게':90}
     ]
cop_df=pd.DataFrame(cop)
cop_df
```

```python
cop_df.to_csv('cop.csv', encoding = 'cp949')
cop_df.to_csv('cop_n.csv', encoding = 'cp949', index = False)
co.to_csv('co.csv', encoding='cp949')
co.to_csv('co_n.csv', encoding='cp949', index = False)
```
